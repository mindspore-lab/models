## 批量压缩
step1: 将matlab生成的数据批量压缩为zip  
step2: 将zip导入到mindspore中

```python
import os
import zipfile

def zip_folder(folder_path, zip_file_path):
    with zipfile.ZipFile(zip_file_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                relative_path = os.path.relpath(file_path, folder_path)
                zipf.write(file_path, relative_path)

def zip_each_folder(target_dir):
    for item in os.listdir(target_dir):
        folder_path = os.path.join(target_dir, item)

        if os.path.isdir(folder_path):
            zip_file_name = item + '.zip'
            zip_file_path = os.path.join(target_dir, zip_file_name)

            zip_folder(folder_path, zip_file_path)

            print(f"{item} 已压缩为 {zip_file_name}")

target_directory = './tryzip'
zip_each_folder(target_directory)
```

## 解压文件
step3: 将存储的按人头分的zip文件，解压到savepath

训练数据

```python
import zipfile
import os


def unzip(zipath, savefolder):
    '''
    zipath     : 待解压文件的路径
    savefolder : 解压后文件存放的文件夹的绝对路径
    '''
    zf = zipfile.ZipFile(zipath)  # zipfile 读取压缩文件对象
    zf.extractall(savefolder)     # 压缩文件内全部文件解压到输入的文件夹中
    zf.close()   # 关闭 zipfile 对象

zipath = os.path.join(os.getcwd(),'1_AP_zip')
savepath = os.path.join(os.getcwd(),'unzip')
os.makedirs(savepath)

fileNames = os.listdir(zipath) 
for file in fileNames:
    index = file.index(".")
    result = file[:index]
    
    filePath = os.path.join(zipath, file)
    savefolder = os.path.join(savepath, result)
    os.makedirs(savefolder)
    unzip(filePath, savefolder)
```

## 转移文件
step4:将解压缩的、按人头分的数据，转换为按类别分的data数据集

针对有类别的数据集

```python
import os
import shutil

def CreateDir(path):
    isExists=os.path.exists(path)
    # 判断结果
    if not isExists:
        # 如果不存在则创建目录
        os.makedirs(path) 
        print(path+' 目录创建成功')
    else:
        # 如果目录存在则不创建，并提示目录已存在
        print(path+' 目录已存在')


def CopyFile(filepath, newPath):
    # 获取当前路径下的文件名，返回List
    fileNames = os.listdir(filepath) 
    for file in fileNames:
        # 将文件命加入到当前文件路径后面
        newDir = os.path.join(filepath,file)
        classes = os.listdir(newDir) 
        for typefile in classes:
            newDirson = os.path.join(newDir,typefile)
            images = os.listdir(newDirson) 
            for image in images:
                newFile = os.path.join(newPath,typefile)
                imagePath = os.path.join(newDirson,image)
                shutil.copy(imagePath, newFile)                            

if __name__ == "__main__":

    rootpath = os.getcwd()
    
    frompath = os.path.join(rootpath,'unzip')
    topath = os.path.join(rootpath, 'data_Sas_AP')
    toApath = os.path.join(topath, 'A')
    toNpath = os.path.join(topath, 'N')
    # 创建目标文件夹
    CreateDir(topath)
    CreateDir(toApath)
    CreateDir(toNpath)
    CopyFile(frompath, topath)
    shutil.rmtree(frompath) 
```

## 拆分文件
将训练数据划分为train和test

```python
import os
from shutil import copy, rmtree
import random
import json


def mk_file(file_path: str):
    if os.path.exists(file_path):
        # 如果文件夹存在，则先删除原文件夹在重新创建
        rmtree(file_path)
    os.makedirs(file_path)


def main():
    # 保证随机可复现
    random.seed(0)

    # 将数据集中10%的数据划分到验证集中
    split_rate = 0.1

    # 指向你解压后的flower_photos文件夹
    cwd = os.getcwd()
    data_root = os.path.join(cwd, "data_Sas_AP")
    assert os.path.exists(data_root), "path '{}' does not exist.".format(data_root)

    flower_class = [cla for cla in os.listdir(data_root)
                    if os.path.isdir(os.path.join(data_root, cla))]

    # 排序，保证顺序一致
    flower_class.sort()
    # # 生成类别名称以及对应的数字索引
    # class_indices = dict((k, v) for v, k in enumerate(flower_class))
    # json_str = json.dumps(dict((val, key) for key, val in class_indices.items()), indent=4)
    # with open('class_indices.json', 'w') as json_file:
    #     json_file.write(json_str)

    cwd = os.getcwd() #返回当前工作目录
    learn_root = os.path.join(cwd,"model_generator_sas")

    # 建立保存训练集的文件夹
    train_root = os.path.join(learn_root, "train")
    mk_file(train_root)
    for cla in flower_class:
        # 建立每个类别对应的文件夹
        mk_file(os.path.join(train_root, cla))

    # 建立保存测试集的文件夹
    test_root = os.path.join(learn_root, "test")
    mk_file(test_root)
    for cla in flower_class:
        # 建立每个类别对应的文件夹
        mk_file(os.path.join(test_root, cla))


    for cla in flower_class: # cla:A/N
        cla_path = os.path.join(data_root, cla) 
        images = os.listdir(cla_path)
        num = len(images)
        # 随机采样测试&验证集的索引
        eval_index = random.sample(images, k=int(num*split_rate))
        for index, image in enumerate(images):
            if image in eval_index:
                # 将分配至测试&测试集中的文件复制到相应目录
                image_path = os.path.join(data_root, cla, image)
                new_path = os.path.join(test_root, cla)
                copy(image_path, new_path)
            else:
                # 将分配至训练集中的文件复制到相应目录
                image_path = os.path.join(data_root, cla, image)
                new_path = os.path.join(train_root, cla)
                copy(image_path, new_path)
            print("\r[{}] processing [{}/{}]".format(cla, index+1, num), end="")  # processing bar
    print("Finish!")
    
if __name__ == '__main__':
    main()
```

## 复制文件
将test内的文件复制到train中，最终使用train集进行模型训练

```python
import os
import shutil

def copy_folder(source_folder, destination_folder):
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    for item in os.listdir(source_folder):
        source = os.path.join(source_folder, item)
        destination = os.path.join(destination_folder, item)

        if os.path.isdir(source):
            copy_folder(source, destination)
        else:
            shutil.copy2(source, destination)

# 使用示例
source = './model_generator_sas/test'
destination = './model_generator_sas/train'

copy_folder(source, destination)
```
